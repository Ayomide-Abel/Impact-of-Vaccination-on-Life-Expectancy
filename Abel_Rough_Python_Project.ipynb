```python
import pandas as pd
```

```python
# Load datasets
vaccination_data = pd.read_csv('global-vaccination-coverage/global-vaccination-coverage.csv')
life_expectancy_data = pd.read_csv('life-expectancy/life-expectancy.csv')
vaccination_data
```

```python
# Display the first few rows of each dataset
print(vaccination_data.head())
print(life_expectancy_data.head())

# Check column names and data types
print(vaccination_data.info())
print(life_expectancy_data.info())
```

```python
# Display first few rows
print("Vaccination Data:")
print(vaccination_data.head())

print("\nLife Expectancy Data:")
print(life_expectancy_data.head())
```

```python
# Display column names for verification
print("\nVaccination Data Columns:")
print(vaccination_data.columns)

print("\nLife Expectancy Data Columns:")
print(life_expectancy_data.columns)
```

```python
# Rename multiple columns in one dictionary
life_expectancy_data.rename(columns={
    'Period life expectancy at birth - Sex: all - Age: 0': 'Life Expectancy',
    'Entity': 'Country'
}, inplace=True)
vaccination_data.rename(columns={'Entity': 'Country'}, inplace=True)

# Verify the column rename
print(life_expectancy_data.columns)
print("\nUpdated Vaccination Data Columns:")
print(vaccination_data.columns)

```

```python
# Select relevant columns for vaccination data
vaccination_columns = ['Country', 'Year', 
                       'BCG (% of one-year-olds immunized)', 
                       'HepB3 (% of one-year-olds immunized)', 
                       'Hib3 (% of one-year-olds immunized)', 
                       'IPV1 (% of one-year-olds immunized)', 
                       'MCV1 (% of one-year-olds immunized)', 
                       'PCV3 (% of one-year-olds immunized)', 
                       'Pol3 (% of one-year-olds immunized)', 
                       'RCV1 (% of one-year-olds immunized)', 
                       'RotaC (% of one-year-olds immunized)', 
                       'YFV (% of one-year-olds immunized)', 
                       'DTP3 (% of one-year-olds immunized)']
vaccination_data = vaccination_data[vaccination_columns]

# Select relevant columns for life expectancy data
life_expectancy_data = life_expectancy_data[['Country', 'Year', 'Life Expectancy']]
```

```python
# Check for missing values in both datasets
print("Missing Values in Vaccination Data:")
print(vaccination_data.isnull().sum())

print("\nMissing Values in Life Expectancy Data:")
print(life_expectancy_data.isnull().sum())

# Drop rows with missing values
vaccination_data.dropna(inplace=True)
life_expectancy_data.dropna(inplace=True)

# Verify after dropping
print("\nVaccination Data after dropping missing values:")
print(vaccination_data.info())

print("\nLife Expectancy Data after dropping missing values:")
print(life_expectancy_data.info())
```

```python
# Merge the datasets on Entity and Year
merged_data = pd.merge(vaccination_data, life_expectancy_data, on=['Country', 'Year'])

# Display the first few rows of the merged dataset
print("Merged Dataset:")
print(merged_data.head())
```

```python
# Summary statistics
print("Summary Statistics:")
print(merged_data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Scatter plot: Vaccination Rate (e.g., DTP3) vs Life Expectancy
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_data, x='DTP3 (% of one-year-olds immunized)', y='Life Expectancy')
plt.title('DTP3 Coverage vs Life Expectancy')
plt.xlabel('DTP3 Coverage (%)')
plt.ylabel('Life Expectancy (years)')
plt.show()

```

```python
# Line plot: Trends of BCG over the years
plt.figure(figsize=(14, 8))
sns.lineplot(data=merged_data, x='Year', y='BCG (% of one-year-olds immunized)', hue='Country')
plt.title('BCG Vaccination Coverage Trends by Country')
plt.xlabel('Year')
plt.ylabel('BCG Coverage (%)')
plt.show()
```
